```python
#1546

N = int(input())
result = input().split()
avg = 0
result_lst = [0 for _ in range(N)]

for i in range(len(result)):
    result_lst[i] = int(result[i])

max_score = max(result_lst)

for i in range(len(result_lst)):
    result_lst[i] = result_lst[i]/max_score * 100

    avg += result_lst[i]
print(avg/N)
```

```python
#4344

n = int(input())

for i in range(n):
    avg = 0
    avg_lst = []
    case = list(map(int, input().split()))

    # for j in range(1, len(case)):
    #     avg += case[j]
    # avg /= case[0]

    avg = sum(case[1:]) / case[0]

    for j in range(1, len(case)):
        if case[j]>avg:
            avg_lst.append(case[j])

    print(f'{len(avg_lst)/case[0]*100:.3f}%')


```

```python
#11728

N, M = map(int, input().split())
i=0
j=0
A = []
B = []
result = []

A_elem = list(map(int, input().split()))

for a in range(len(A_elem)):
    A.append(A_elem[a])

B_elem = list(map(int, input().split()))

for b in range(len(B_elem)):
    B.append(B_elem[b])

while(i<N and j<M):
    if(A[i] < B[j]):
        result.append(A[i])
        i+=1
    else:
        result.append(B[j])
        j+=1


for index in range(i, N):
    result.append(A[index])

for index in range(j, M):
    result.append(B[index])

print(result)
```

```python
#9012

def foo(ps):
    stack = []
    for i in ps:
        if i=='(':
            stack.append(i)
        else:
            if len(stack) == 0:
                print("NO")
                return
            else:
                del stack[len(stack)-1]
    if len(stack) == 0:
        print("YES")
    else:
        print("NO")

n = int(input())

for _ in range(n):
    PS = list(input())
    foo(PS)


```
