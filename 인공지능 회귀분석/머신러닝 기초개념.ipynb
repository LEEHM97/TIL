#### Loss Function

* Loss Function = Target Function
* Loss = y - y' = (target value) - (predicted value)
* y - y' = L = wx + b

    -> x,y는 상수이므로 고정된 값. 

    -> w와 b는 parameter(=weight)로 이를 조절하여 L을 0으로 만들어준다

    -> **따라서 L은 w에 대한 함수**


* Loss가 0이다 --> (y = y')
