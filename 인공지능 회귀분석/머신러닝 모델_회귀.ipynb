## 회귀란?

* 주어진 데이터(X)와 원하는 값(y) 사이의 관계를 찾는 방법
* 주어진 데이터(X)를 통해서 원하는 값(y = target value)을 예측하는 방법
    - 특정 값을 예측하고 찾는 문제는 회귀를 사용한다.
    - e.g. 주식, 일치율, 집값 예측 등등


## 회귀 모델의 종류

### 1) Linear Regression
image.png

* 가장 직관적이고 많이 사용되는 선형 회귀 모델

* 이차항 이상이 존재하지 않는 linear model

* e.g. 𝘧: 𝒲1𝒳1 + 𝒲2𝒳2 + b = $\hat{y}$
    > 함수 𝘧가 모든 점을 지나게 할 수 없으니 오차가 가장 적은 𝘧를 구해야한다.
    >
    > MSE가 작아지는 Parameter(𝒲1, 𝒲2, b)를 구한다.
    >
    > b가 없다면 𝘧가 무조건 원점을 지나야 하므로 자유도가 떨어진다.

* 𝘧가 비선형(polynomial)이면 $P(train)\,⇑,\;\;P(test)\,↓ \;\;$ : overfitting

* 𝘧가 선형(linear)이면 $P(train)\,$ ↑ $,\;\;P(test)\,↓ \;\;$ : ideal

* 실제 데이터 분석 대회에서 가장 많이 사용하는 효과적인 회귀 모델

## 회귀 모델 평가_대표적 회귀 모델 평가 지표

* 주어진 데이터로 모델을 학습시키는 것은 **지정한 성능 평가 지표를 향상시키는 과정**이다.

* 성능 평가 지표의 값은 **"예측 성능"**을 기준으로 한다.

* 정량적 기준을 설정하고, 달성할 때까지 모델을 학습시키고 성능을 개선한다.

* 목표한 성능에 도달한 모델을 실제 서비스에 적용


### 1. MSE(Mean Square Error)


* $\frac{\sum_{i}^n(yᵢ-\hat{y}ᵢ)^2}{N}$
    > $yᵢ-\hat{y}ᵢ$가 음수, 양수로 번갈아가며 나오면 오차가 상쇄되기 때문에 제곱을 시켜준다.
    >
    > $yᵢ-\hat{y}ᵢ$를 제곱해주면 데이터가 많아질수록 loss가 커지기 때문에 N으로 나누어 평균으로 정의한다.

### 2. RMSLE(Root Mean Squared Log Error)

* $\sqrt\frac{\sum_{i}^n(yᵢ-\hat{y}ᵢ)^2}{N}$
    > $yᵢ$와 $\hat{y}ᵢ$의 scale 영향을 받지 않는다.
    >
    > outlier에 robust한다(영향을 덜 받는다)

### 3. MAE(Mean Absolute Error)

* $\frac{\sum_{i}^n\lvert yᵢ-\hat{y}ᵢ\lvert}{N}$


### 4. $R^2$ Score(Coefficient of Determination)

* 1 - $\frac{SSE}{SST}$ (=$\frac{MSE}{y의 분산}\;\;\;$)
    > 0~1사이값 (1일수록 선형적 모델링)
